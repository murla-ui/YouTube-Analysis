# 4.3 Analysis by Category

## Which video category has the largest number of trending videos?

First, we will add a column that contains category names based on the values in category_id column. We will use a category JSON file provided.

```python
import json

with open("US_category_id.json") as f:
    categories = json.load(f)["items"]
cat_dict = {}
for cat in categories:
    cat_dict[int(cat["id"])] = cat["snippet"]["title"]
combined_csv['category_name'] = combined_csv['category_id'].map(cat_dict)
```

```python
combined_csv.head()
```

```python
cdf = combined_csv["category_name"].value_counts().to_frame().reset_index()
cdf.rename(columns={"index": "category_name", "category_name": "No_of_videos"}, inplace=True)
fig, ax = plt.subplots(figsize=(10,6))
_ = sns.barplot(x="category_name", y="No_of_videos", data=cdf)
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
_ = ax.set(xlabel="Category", ylabel="No. of videos")
plt.title("Categories of trending videos by No. of videos")
```

We see that the Entertainment category contains the largest number of trending videos among other categories: around 65 000 videos, followed by People&Blogs category with around 20 000 videos, followed by Music category with around 20 000 videos, and so on.

# 4.5 Likes/dislikes and views/comments ratio trending videos have by category

```python
like_dislike_ratio = combined_csv.groupby('category_name')['likes'].agg('sum') / combined_csv.groupby('category_name')['dislikes'].agg('sum')
like_dislike_ratio = like_dislike_ratio.sort_values(ascending=False).reset_index()
like_dislike_ratio.columns = ['category_name','ratio']
plt.subplots(figsize=(10, 15))
sns.barplot(x="ratio", y="category_name", data=like_dislike_ratio,
            label="Likes/dislikes Ratio")

plt.title("Likes/dislikes ratio by category")
```

```python
views_comments_ratio = combined_csv.groupby('category_name')['views'].agg('sum') / combined_csv.groupby('category_name')['comment_count'].agg('sum')
views_comments_ratio = views_comments_ratio.sort_values(ascending=False).reset_index()
views_comments_ratio.columns = ['category_name','ratio']
plt.subplots(figsize=(10, 15))
sns.barplot(x="ratio", y="category_name", data=views_comments_ratio,
            label="Views/comments Ratio")

plt.title("Views/comments ratio by category")
```

- Pets & Animals videos have highest likes-dislikes ratio. Relatively most liked videos.
- Nonprofit & Activism's videos have lowest likes-dislike ratio and views-comments ratio. People don't like these video and comment there a lot.
- Movies is a category people rarely comment compatatively with views.

# 4.6 When was videos published?

```python
import datetime

combined_csv['publish_month'] = combined_csv['publish_time']. apply(
    lambda x: datetime.datetime.strptime(x[:10], "%Y-%m-%d").date().strftime('%B'))
```

Using datetime library we are adding new column 'publish_month' into the dataset

```python
combined_csv.head()
```

```python
mdf = combined_csv["publish_month"].value_counts().to_frame().reset_index()
mdf.rename(columns={"index": "publish_month", "publish_month": "No_of_videos"}, inplace=True)
fig, ax = plt.subplots(figsize=(10,6))
_ = sns.barplot(x="publish_month", y="No_of_videos", data=mdf)
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
_ = ax.set(xlabel="Month", ylabel="No. of videos")
```

According to this barchart, most popular month for trending videos id May, December and March, least popular September, August and July
